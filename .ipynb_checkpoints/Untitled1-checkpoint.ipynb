```python
## import packages

from bs4 import BeautifulSoup
import requests
import pandas as pd
from datetime import datetime
```

```python
url = 'https://tw.appledaily.com/search/result?sort=time&searchType=all&dateStart=2011%2F11%2F21&dateEnd=2012%2F01%2F13&querystrS=%E9%A6%AC%E8%8B%B1%E4%B9%9D'
r = requests.get(url) #抓取網址
web_content = r.text #網址的text
soup = BeautifulSoup(web_content, 'lxml')
```

```python
soup
```

```python
title = soup.find_all('div', class_ = 'content') # 搜索新聞的位置
#print(title)

#取得新聞標題
headers = [t.find('a').text for t in title]
#print(titles)
headers = [' '.join(t.strip()) for t in headers]
headers
```

```python
# extract url from title
test = [t.find('a') for t in title]
for t in test:
    print(type(t))
```

```python
news_url = []
for t in test:
    news_url.append(t.get('href'))
news_url
len(news_url)
len(headers)
```

```python
p = [p.find('p') for p in title]
article = []
for content in p:
    article.append(content.text)
article
```

```python
time = [t.find('time').text for t in title]
#time = [t.text for t in time]

time_dt = []
for i in time:
    time_dt.append(datetime.strptime(i, '%Y%m%d'))
time_dt

media = []
for i in range(10):
    media.append('Apple Daily')
len(media)


df_new = pd.DataFrame(
{
    'titles' : headers,
    'links' : news_url,
    'content' : article,
    'time' : time_dt,
    'media' : media
})



df_new.to_csv('/Users/garday/Documents/MY499/Data Collection/Data/馬英九_2012news.csv', \
             encoding = 'utf_8_sig', index = None)
```

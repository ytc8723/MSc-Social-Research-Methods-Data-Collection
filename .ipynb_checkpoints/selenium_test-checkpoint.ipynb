```python
headers = {
    'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.106 Safari/537.36',
    'Cookie' : 'nxtu=1591260170.0816D0DB3A-AB37-47ED-88F3-B6C2FCC98798; _ga=GA1.2.462608449.1591260170; __gads=ID=11850eba1a242aa8:T=1591260184:S=ALNI_MabTpTOWv5gzT0dhbZh-amJHl2X7g; nxtu=1591260170.0816D0DB3A-AB37-47ED-88F3-B6C2FCC98798; GDPR_ACCP=Y; akaas_TWOrigin01=2147483647~rv=4~id=b2a217b73529b7f2b291d9b87e669979; DatalayerGeo=%7B%22ipinfo%22%3A%7B%22ip_address%22%3A%2261.219.175.56%22%2C%22ip_type%22%3A%22null%22%2C%22Network%22%3A%7B%22organization%22%3A%22null%22%2C%22carrier%22%3A%22null%22%2C%22asn%22%3A%22null%22%2C%22connection_type%22%3A%22null%22%2C%22line_speed%22%3A%22null%22%2C%22ip_routing_type%22%3A%22null%22%2C%22Domain%22%3A%7B%22tld%22%3A%22null%22%2C%22sld%22%3A%22null%22%7D%7D%2C%22Location%22%3A%7B%22continent%22%3A%22null%22%2C%22latitude%22%3A%2225.047760%22%2C%22longitude%22%3A%22121.531853%22%2C%22CountryData%22%3A%7B%22country%22%3A%22Taiwan%2C%20Province%20of%20China%22%2C%22country_code%22%3A%22TW%22%2C%22country_cf%22%3A%22null%22%2C%22region%22%3A%22Taipei%22%2C%22StateData%22%3A%7B%22state%22%3A%22Taipei%22%2C%22state_code%22%3A%22null%22%2C%22state_cf%22%3A%22null%22%7D%2C%22dma%22%3A%22null%22%2C%22msa%22%3A%22null%22%2C%22CityData%22%3A%7B%22city%22%3A%22Taipei%22%2C%22postal_code%22%3A%2210048%22%2C%22time_zone%22%3A%22%2B08%3A00%22%2C%22area_code%22%3A%22null%22%2C%22city_cf%22%3A%22null%22%7D%7D%7D%7D%7D; isLoggedIn=false; _gid=GA1.2.743474847.1592554598; RT="z=1&dm=appledaily.com&si=29n6yreuyof&ss=kbkh9kw1&sl=0&tt=0"; country=TW; page_id=1592554787.961-b294-098b-e187-666fd29f87d1; adisblk=N'
}
url = 'https://tw.appledaily.com/search/result?sort=time&searchType=all&dateStart=2011%2F11%2F21&dateEnd=2012%2F01%2F13&querystrS=%E9%A6%AC%E8%8B%B1%E4%B9%9D'
browser = webdriver.Chrome('/Users/garday/Downloads/chromedriver')
browser.get(url)
i = 0
for i in range(5):
    browser.find_element_by_id('moreButton').click()
    time.sleep(5)
    soup = BeautifulSoup(browser.page_source, 'lxml')

    
title = soup.find_all('div', class_ = 'content') # 搜索新聞的位置
# print(title)

#取得新聞標題
headers = [t.find('a').text for t in title]
#print(titles)
headers = [' '.join(t.strip()) for t in headers]
headers

# 取得新聞連結
test = [t.find('a') for t in title]
news_url = []
for t in test:
    news_url.append(t.get('href'))
news_url

# 取得新聞內文
p = [p.find('p') for p in title]
article = []
for content in p:
    article.append(content.text)
article 

# 取得發佈時間
time = [t.find('time').text for t in title]
#time = [t.text for t in time]
time_dt = []
for i in time:
    time_dt.append(datetime.strptime(i, '%Y%m%d'))
    
```

```python
browser.close()
```

```python
df_new = pd.DataFrame(
{
    'titles' : headers,
    'links' : news_url,
    'content' : article,
    'time' : time_dt,
})

df_new.to_csv('/Users/garday/Documents/MY499/Data Collection/Data/馬英九_2012news.csv', \
             encoding = 'utf_8_sig', index = None)
```

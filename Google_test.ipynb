```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import scrapy
import re
from lxml import etree
```

```python
url = 'https://news.google.com/search?q=%E9%81%B8%E8%88%89&hl=zh-TW&gl=TW&ceid=TW%3Azh-Hant'

r = requests.get(url) #抓取網址
web_content = r.text #網址的text
soup = BeautifulSoup(web_content, 'lxml')
```

```python
title = soup.find_all('div', class_ = 'xrnccd') # xrnccd Google新聞的分隔區
titles = [t.find('span').text for t in title]
titles
```

```python
news_url = [requests.get(t.find('a')['href'].replace('.', 'https://news.google.com',1)).url for t in title]
news_url
```

```python
# create data frame 
df = pd.DataFrame(
{
    'title' : titles,
    'links' : news_url
})

df.to_csv('/Users/garday/Documents/MY499/Data Collection/Data/news_選舉.csv', \
         encoding = 'utf_8_sig', index = None)
```

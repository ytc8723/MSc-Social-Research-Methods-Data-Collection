```python
# 蔡英文2020
url_t_2020 = 'https://tw.appledaily.com/search/result?sort=time&searchType=all&dateStart=2019%2F11%2F18&dateEnd=2020%2F01%2F10&querystrS=%E8%94%A1%E8%8B%B1%E6%96%87'
browser = webdriver.Chrome('/Users/garday/Downloads/chromedriver')
browser.get(url_t_2020)
i = 0
for i in range(211):
    browser.find_element_by_id('moreButton').click()
    time.sleep(2)
    soup_c = BeautifulSoup(browser.page_source, 'lxml')
```

```python
def extract(soup):
    title = soup.find_all('div', class_ = 'content')
    #取得新聞標題
    headers_c = [t.find('a').text for t in title]
    #print(titles)
    headers_c = [' '.join(t.strip()) for t in headers_c]
    headers_c

    # 取得新聞連結
    test_c = [t.find('a') for t in title]
    news_url_c = []
    for t in test_c:
        news_url_c.append(t.get('href'))
    news_url_c

    # 取得新聞內文
    p_c = [p.find('p') for p in title]
    article_c = []
    for content in p_c:
        article_c.append(content.text)
    article_c

    # 取得發佈時間
    time_text_c = [t.find('time').text for t in title]
    #time = [t.text for t in time]
    time_dt_c = []
    for i in time_text_c:
        time_dt_c.append(datetime.strptime(i, '%Y%m%d'))

    media_c = []
    for i in range(len(time_dt_c)):
        media_c.append('Apple Daily')
        
    return headers_c, news_url_c, article_c, time_dt_c, media_c

```

```python
result = extract(soup_c)
headers = result[0]
news_url = result[1]
article = result[2]
time_dt = result[3]
media = result[4]
headers
```

```python
df_tyw_2020 = pd.DataFrame(
{
    'titles' : headers,
    'links' : news_url,
    'content' : article,
    'time' : time_dt,
    'media': media
})

df_tyw_2020.drop_duplicates('titles', 'first', inplace = True)

df_tyw_2020.to_csv('/Users/garday/Documents/MY499/Data Collection/Data/蔡英文_2020_applenews.csv', \
             encoding = 'utf_8_sig', index = None)
```

```python
# 馬英九2012
url = 'https://news.ltn.com.tw/search?keyword=%E9%A6%AC%E8%8B%B1%E4%B9%9D&conditions=and&start_time=2011-11-21&end_time=2012-01-13&page='
headers = []
news_url = []
article = []
time_dt = []
for page_num in range(1,122):
    res_news = requests.get(url + repr(page_num))
    newslist_soup = BeautifulSoup(res_news.text, 'lxml')
    
    for newslist in newslist_soup.find_all('div', class_ = 'whitecon'):
        li_label = newslist.find_all('li')
        
      # 新聞標題
    for t in li_label:
        headers.append(t.find('a').text)
    
    # 新聞連結
    url_sect = [t.find('a') for t in li_label]
    for t in url_sect:
        news_url.append(t.get('href'))
    
    # 新聞abstract
    p = [p.find('p') for p in li_label]

    for content in p: 
        article.append(content.text)
    
    # 時間
    for t in li_label:
        time_dt.append(t.find('span').text)

    time.sleep(3)
    
    
        
        
                
```

```python
media = []
for i in range(len(time_dt)):
    media.append('Liberal Times')

df = pd.DataFrame(
{
    'titles' : headers,
    'links' : news_url,
    'content' : article,
    'time' : time_dt,
    'media': media
})

df.drop_duplicates('titles', 'first', inplace = True)
df

# df.to_csv('/Users/garday/Documents/MY499/Data Collection/Data/馬英九_2012_liberalnews.csv', \
#              encoding = 'utf_8_sig', index = None)
```

```python
# 蔡英文2012 
url_t = 'https://news.ltn.com.tw/search?keyword=%E8%94%A1%E8%8B%B1%E6%96%87&conditions=and&start_time=2011-11-21&end_time=2012-01-13&page='

def liberal_times_webcrawl(url):
    
    headers = []
    news_url = []
    article = []
    time_dt = []
    for page_num in range(1,117):
        res_news = requests.get(url + repr(page_num))
        newslist_soup = BeautifulSoup(res_news.text, 'lxml')

        for newslist in newslist_soup.find_all('div', class_ = 'whitecon'):
            li_label = newslist.find_all('li')

          # 新聞標題
        for t in li_label:
            headers.append(t.find('a').text)

        # 新聞連結
        url_sect = [t.find('a') for t in li_label]
        for t in url_sect:
            news_url.append(t.get('href'))

        # 新聞abstract
        p = [p.find('p') for p in li_label]

        for content in p: 
            article.append(content.text)

        # 時間
        for t in li_label:
            time_dt.append(t.find('span').text)

        time.sleep(2)
    
    media = []
    for i in range(len(time_dt)):
        media.append('Liberal Times')
        
    return headers, news_url, article, time_dt, media 
```

```python
tsy_2012 = liberal_times_webcrawl(url_t)
```

```python
df_tsy_2012 = pd.DataFrame({
    'titles' : tsy_2012[0],
    'links' : tsy_2012[1],
    'content' : tsy_2012[2],
    'time' : tsy_2012[3],
    'media': tsy_2012[4]
})
df_tsy_2012.drop_duplicates('titles', 'first', inplace = True)

df_tsy_2012.to_csv('/Users/garday/Documents/MY499/Data Collection/Data/2012/蔡英文_2012_liberal.csv', \
                  encoding = 'utf_8_sig', index = None)
```
